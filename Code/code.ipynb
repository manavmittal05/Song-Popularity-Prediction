```python
# label encode 

from sklearn.preprocessing import LabelEncoder

model = LabelEncoder()

df['label'] = model.fit_transform(df['label'])

df.head()
```

```python
target = df['label']
df = df.drop('label' , axis = 1)

df.head()
```

```python
song_name = df['song_name']
df = df.drop('song_name' , axis = 1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split



x_train , x_test , y_train , y_test  = train_test_split(df , target , test_size= 0.2 , random_state=45)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Decision tree classifier 

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model = DecisionTreeClassifier(max_depth= None , criterion='entropy' , max_features=None)

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

from sklearn.metrics import accuracy_score

accuracy_score(y_test , y_pred)
```

```python
# logistic regression 

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(penalty='l2' , max_iter=1000)

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```

```python
# naive byes

from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```

```python
# random forest 

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=1000, max_depth=None , n_jobs=4)

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```

```python
# ensemble model of weak classifiers 

models = []

models.append(('decision tree' , DecisionTreeClassifier()))
models.append(('logistic regression' , LogisticRegression()))
models.append(('naive byes' , GaussianNB()))

from sklearn.ensemble import VotingClassifier

model = VotingClassifier(estimators=models , voting='soft')

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```

```python
# outlier detection using LOF 

from sklearn.neighbors import LocalOutlierFactor

df['target'] = target 

model = LocalOutlierFactor(n_neighbors=20 , contamination=0.2)

y_pred = model.fit_predict(df)

y_pred

df['outlier'] = y_pred

df.head()
```

```python
# drop outliers 

df = df[df['outlier'] != -1]
```

```python
df.shape
```

```python
# train test split 

target = df['target']
df = df.drop('target' , axis = 1)

x_train , x_test , y_train , y_test  = train_test_split(df , target , test_size= 0.2 , random_state=42)
```

```python
# use logistic regression 

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(penalty='l2' , max_iter=1000)

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```

```python
# random forest 

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=1000, max_depth=None , n_jobs=4)

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```

```python
# adaboost 

from sklearn.ensemble import AdaBoostClassifier

model = AdaBoostClassifier(n_estimators=1000 , learning_rate=0.01)

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```

```python
# KNN

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=1 , n_jobs=4)

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```
