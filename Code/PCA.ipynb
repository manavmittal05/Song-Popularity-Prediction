```python
# plot graph of silhouette score vs k 

from sklearn.cluster import KMeans

from sklearn.metrics import silhouette_score

max_silhouette = -1 
max_k = -1 

silhouette = []
elbow_score = []

for k in range(2, 30):
    kmeans = KMeans(n_clusters = k).fit(df)
    s_score = silhouette_score(df, kmeans.labels_)
    silhouette.append(s_score)
    elbow_score.append(kmeans.inertia_)
    if s_score > max_silhouette: 
        max_k = k 
        max_silhouette = s_score


plt.plot(range(2, 30), silhouette)
plt.title('Silhouette')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()

plt.plot(range(2, 30), elbow_score)
plt.title('Elbow Score')
plt.xlabel('Number of clusters')
plt.ylabel('Elbow Score')
plt.show()
```

```python
# logistic regression 

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(penalty='l2' , max_iter=1000)

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```

```python
# ensemble model of weak classifiers 

models = []

models.append(('decision tree' , DecisionTreeClassifier()))
models.append(('logistic regression' , LogisticRegression()))
models.append(('naive byes' , GaussianNB()))

from sklearn.ensemble import VotingClassifier

model = VotingClassifier(estimators=models , voting='soft')

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```

```python
# outlier detection using LOF 

from sklearn.neighbors import LocalOutlierFactor

df['target'] = target 

model = LocalOutlierFactor(n_neighbors=20 , contamination=0.2)

y_pred = model.fit_predict(df)

y_pred

df['outlier'] = y_pred

df.head()
```

```python
# drop outliers 

df = df[df['outlier'] != -1]
```

```python
df.shape
```

```python
# train test split 

target = df['target']
df = df.drop('target' , axis = 1)
df = df.drop('outlier' , axis = 1)

x_train , x_test , y_train , y_test  = train_test_split(df , target , test_size= 0.1 , random_state=42)
```

```python
# use logistic regression 

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(penalty='l2' , max_iter=1000)

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```

```python
# random forest 

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=1000, max_depth=None , n_jobs=4)

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```

```python
# adaboost 

from sklearn.ensemble import AdaBoostClassifier

model = AdaBoostClassifier(n_estimators=1000 , learning_rate=0.01)

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```

```python
# KNN

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=1 , n_jobs=4)

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```

```python
# do random forect using feature bagging and sample bagging 

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=1000, max_depth=None , n_jobs=4 , max_features=0.8 , bootstrap=True)

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```

```python
# random forest 

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=1000, max_depth=None , n_jobs=-1)

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test , y_pred)
```
